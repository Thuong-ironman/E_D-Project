```python
#https://tesseract-ocr.github.io/tessdoc/Installation.html
import pytesseract, cv2, glob, os
import numpy as np

#https://stackoverflow.com/questions/50951955/pytesseract-tesseractnotfound-error-tesseract-is-not-installed-or-its-not-i
pytesseract.pytesseract.tesseract_cmd = r'C:/Program Files/Tesseract-OCR/tesseract.exe'
```

```python
def preprocess(image):
    #scale up image
    scale_percent = 150 # percent of original size
    width = int(image.shape[1] * scale_percent / 100)
    height = int(image.shape[0] * scale_percent / 100)
    dim = (width, height)
    image = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)

    #binarization
    #image = cv2.threshold(image,200,255,cv2.THRESH_BINARY)[1]
    image = cv2.medianBlur(image, 3)
    image = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,35,5)
    kernel = np.ones((2, 2), np.uint8)
    image = cv2.dilate(image, kernel, iterations=1)
    return image

def get_tesseract_preds(input_dir):
    for img_path in glob.glob(os.path.join(input_dir, '*.png')):
        # load the input image and convert it from BGR to RGB channel ordering
        image = cv2.imread(img_path, 0)
        #image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # preprocess
        image = preprocess(image)
        cv2.imshow(f'{img_path}', image)
        # use Tesseract to OCR the image, adding custom options
        custom_config = r'--oem 3 --psm 6'
        #s = pytesseract.image_to_string(image, config=custom_config)
        d = pytesseract.image_to_data(image, output_type=pytesseract.Output.DICT)
        print(d.keys())
        n_boxes = len(d['level'])
        for i in range(n_boxes):
            if(len(d['text'][i])):
                print(f'Box {i+1}: {d["text"][i]}')
                (x, y, w, h) = (d['left'][i], d['top'][i], d['width'][i], d['height'][i])
                cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 1)
                cv2.putText(image, f'{i+1}', (x,y), fontFace=cv2.FONT_HERSHEY_TRIPLEX, fontScale=0.4, color=(0,0,0),thickness=1)

        cv2.imshow(f'{img_path}', image)
        cv2.waitKey()
        cv2.destroyAllWindows()
    
```
