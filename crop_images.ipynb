```python
import cv2 as cv
import glob, os, math
```

```python
data1_dir = 'Data/label-project/data_1/'
data2_dir = 'Data/label-project/data_2/'
imgs1_dir = data1_dir + 'images/'
imgs2_dir = data2_dir + 'images/'
icdar1_dir = data1_dir + 'icdar/'
icdar2_dir = data2_dir + 'icdar/'
datasets = [[imgs1_dir, icdar1_dir, data1_dir], [imgs2_dir, icdar2_dir, data2_dir]]
```

```python
# loads all images from folder, crops according to annotations and converts to grayscale
def preprocess_imgs(imgs_dir, annos_dir, data_dir):
    cropped_dir = data_dir + 'cropped'
    if not os.path.exists(cropped_dir):
        os.makedirs(cropped_dir)
    imgs = list()
    for dir_path in glob.glob(imgs_dir):
        #img_count = 0
        for img_path in glob.glob(os.path.join(dir_path, '*.png')):
            img_name = img_path.split('\\')[1][:-4]
            anno_path = annos_dir + 'gt_' + img_name + '.txt'
            with open(anno_path) as anno_file:
                rects = [[int(coord.split('.')[0]) for coord in line.split(' ')] for line in anno_file]
                rects = remove_duplicates(rects)
                for i in range(len(rects)):
                    pt1 = (rects[i][0], rects[i][1])
                    pt2 = (rects[i][2], rects[i][3])
                    img = cv.imread(img_path)
                    gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
                    crop_img = gray_img[pt1[1]:pt2[1], pt1[0]:pt2[0]]
                    cv.imwrite(cropped_dir + '/' + img_name + f'_{i}.png', crop_img)

                    #gray_img = cv.resize(gray_img, (1200, 800))
                    #cv.imshow('img', gray_img)
                    #cv.waitKey()
                    #cv.destroyAllWindows()
            #img_count+=1
            #print(f'img {img_count} done')

# kinda sketchy, because of arbitrary interval for dublicity check
def remove_duplicates(rects):
    dupes_idxs = list()
    for i in range(len(rects)):
        for j in range(len(rects)-i-1):
            pt1_rect1 = [rects[i][0], rects[i][1]]
            pt2_rect1 = [rects[i][2], rects[i][3]]
            pt1_rect2 = [rects[j+i+1][0], rects[j+i+1][1]]
            pt2_rect2 = [rects[j+i+1][2], rects[j+i+1][3]]
            if abs(math.dist(pt1_rect1, pt1_rect2)) <= 30 and abs(math.dist(pt2_rect1, pt2_rect2)) <= 30:
                dupes_idxs.append(j+i+1)
    for idx in dupes_idxs[::-1]:
        del rects[idx]
    return rects
```

```python
for dataset in datasets:
    preprocess_imgs(dataset[0], dataset[1], dataset[2])
```
